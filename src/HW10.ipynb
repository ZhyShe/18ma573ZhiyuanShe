```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#-------------
## Parameters
#-------------

rho=0.4
a=0
b=1
dx=.2
alpha=1
n=(b-a)/dx
M=200
#-------------

dt=rho*(dx**2)/alpha

X_i=[a+i*dx for i in range(0,int(n+1))]
T_k=[k*dt for k in range(M+1)]

```

```

def u(a, b, rho, dx, dt, U):
  n,m=U.shape
  for j in range(m):
    U[0,j]=0
    U[n-1,j]=0
  for i in range(n):
    U[i,0]=np.sin((a+i*dx)*np.pi)
  for j in range(1,m):
    for i in range(1,n-1):
      U[i,j]=rho*U[i-1,j-1]+(1-2*rho)*U[i,j-1]+rho*U[i+1,j-1]

U=np.zeros((int(n)+1,M+1))
for i in range(int(n+1)):
  for j in range(M+1):
    U[i,j]=None
```

```
u(a, b, rho, dx, dt, U)
```

```
Z=np.zeros((int(M+1),int(n+1)))
for j in range(Z.shape[0]):
  for i in range(Z.shape[1]):
    Z[j,i]=U[i,j]
```

```
def exact_u(x,t):
  return np.dot(np.exp(-np.dot(np.pi**2,t)),np.sin(np.dot(np.pi,x)))

def erro_function(a, b, n, M, U):
  exa_u=np.zeros((int(n+1),int(M+1)))
  for i in range(int(n+1)):
    for j in range(int(M+1)):
      exa_u[i,j]=exact_u(a+i*dx,j*dt)
  erro=np.abs(U-exa_u)
  return max([max(k) for k in erro])
```

```
print('The L error is %f'%erro_function(a, b, n, M, U))
```

```
T=[1.95, 1.97, 1.98]
Ys=[]
for t in T:
  Ys.append(exact_u(X_i,t)-U[:,int(t/dt)])
  
```

```
y1,y2,y3=Ys
plt.plot(X_i,y1)
plt.plot(X_i,y2)
plt.plot(X_i,y3)
plt.show()
```

```
Z=np.zeros((int(M+1),int(n+1)))
for j in range(Z.shape[0]):
  for i in range(Z.shape[1]):
    Z[j,i]=exact_u(a+i*dx,j*dt)
    
fig = plt.figure()
ax =Axes3D(fig)
X=np.array([a+i*dx for i in range(int(n+1))])
Y=np.array([k*dt for k in range(M+1)])
X, Y=np.meshgrid(X,Y)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='rainbow')
plt.show()
```
